# EPPS NeurIPS 2024 Evaluation Visualization

This notebook parses the output of `evaluator.py` to calculate the Pragmatic Faithfulness Score (PFS) for the three baselines (Zero-Shot, RAG History, Naive Summary) vs the novel Embodied Pragmatic Program Synthesis (EPPS) Architecture leveraging the Common Ground Model (CGM). 

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Load evaluation results
try:
    with open('../data/eval_results.json', 'r') as f:
        results_data = json.load(f)
except FileNotFoundError:
    print("Results file not found. Ensure you have run evaluator.py first.")
    results_data = {}
```

```python
# Process JSON into a DataFrame
plot_data = []
for agent_type, persona_runs in results_data.items():
    for run in persona_runs:
        plot_data.append({
            "Agent": agent_type,
            "Persona": run["persona"],
            "PFS": run["pfs"]
        })
df = pd.DataFrame(plot_data)
```

```python
# Generate Clustered Bar Chart if data exists
if not df.empty:
    # Aggregate means
    df_agg = df.groupby(["Persona", "Agent"])["PFS"].mean().reset_index()
    
    fig, ax = plt.subplots()
    sns.barplot(
        data=df_agg, 
        x="Persona", 
        y="PFS", 
        hue="Agent", 
        palette="viridis", 
        ax=ax
    )
    
    ax.set_title("Pragmatic Faithfulness Score (PFS) by Persona across Architectures", fontsize=16)
    ax.set_ylabel("Mean PFS (%)", fontsize=12)
    ax.set_xlabel("Simulation Persona", fontsize=12)
    ax.set_ylim(0, 100)
    plt.legend(title="Architecture")
    plt.tight_layout()
    plt.savefig("epps_neurips_figure1.png", dpi=300)
    plt.show()
```
